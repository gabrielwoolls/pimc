```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
```

```python
# plot on single graph
rc = matplotlib.rc
rc('figure', figsize=(5.0, 5.0))
rc('axes', linewidth=1.5)
rc('lines', linewidth=1.5)
rc('font', size=14.0)
rc('xtick.major', size=0.0, pad=8.0)
rc('xtick.minor', size=0.0, pad=8.0)
rc('ytick.major', size=6.0, pad=8.0)
rc('ytick.minor', size=3.0, pad=8.0)




with open('worldlines.txt', 'r') as f:
    data = f.read().split('\n\n')[:-1] # split data into blocks, ignore blank line at end
    fig, axs = plt.subplots(1, figsize=(5,5))

    for i, block in enumerate(data):
        coordinates = np.array([list(map(float, line.split())) for line in block.split('\n')])
        # process coordinates here, for example:

        x = coordinates[:,0]
        y = coordinates[:,1]
        color = plt.cm.rainbow(i/len(data)) # sequential rainbow color based on block number
        axs.plot(x, y, color=color)
        axs.scatter(x[0],y[0], marker='o', color='k') # plot where each one starts

axs.set_xlabel("x")
axs.set_ylabel("y")
axs.set_ylim(-12,12)
axs.set_xlim(-12,12)
axs.set_title(f"$n = {i+1}$")
    
plt.tight_layout()
plt.show()


# in the plot below, you can see that the five particles move so as to spread out from each other but sit at the bottom of the potential well.
```
