```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
```

```python
def plot_data(filename, axis):


    with open(filename, 'r') as f:
        data = f.read().split('\n\n')[:-1] # split data into blocks, ignore blank line at end

        for i, block in enumerate(data):
            coordinates = np.array([list(map(float, line.split())) for line in block.split('\n')])
            # process coordinates here, for example:

            x = coordinates[:,0]
            y = coordinates[:,1]
            color = plt.cm.rainbow(i/len(data)) # sequential rainbow color based on block number

            axis.plot(x, y, color=color)
            axis.scatter(x[0],y[0], marker='o', color='k') # plot where each one starts

```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))

plot_data('build/worldlines_serial.txt', ax[0])
ax[0].title.set_text("serial")

plot_data('build/worldlines_parallel.txt', ax[1])
ax[1].title.set_text("parallel")

for i in range(2):
    ax[i].set_xlabel("x")
    ax[i].set_ylabel("y")
    ax[i].set_ylim(-12,12)
    ax[i].set_xlim(-12,12)

plt.suptitle(f"$n = 50$")

plt.tight_layout()
plt.show()

```
